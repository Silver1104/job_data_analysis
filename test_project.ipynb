```python
from datasets import load_dataset
import pandas as pd

df = load_dataset("lukebarousse/data_jobs")['train'].to_pandas()
df
```

```python
import ast
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda skill_list: ast.literal_eval(skill_list) if pd.notna(skill_list) else skill_list)
```

```python
df_skills_separated = df.explode('job_skills')
```

```python
df_skills_separated['job_skills'].value_counts()
```

```python
skills_count = df_skills_separated.groupby(['job_title_short', 'job_skills']).size()
skills_count
```

```python
df_skills_count = skills_count.reset_index(name='skill_count')
df_skills_count.sort_values('skill_count', ascending=False, inplace=True)
df_skills_count
```

```python
job_title = 'Cloud Engineer'
number = 10

df_final = df_skills_count[df_skills_count['job_title_short'] == job_title].head(number).copy()
```

```python
import matplotlib.pyplot as plt
df_final.plot(kind='barh', x='job_skills', y='skill_count')
plt.gca().invert_yaxis()
plt.ylabel('Job role')
plt.legend().set_visible(False)
plt.title(f'Top {number} skills in demand for role {job_title}')
plt.show()
```

```python
df['job_posted_month'] = df['job_posted_date'].dt.month
df
```

```python
df_ds = df[df['job_title_short'] == 'Data Scientist'].copy()
df_ds_explode = df_ds.explode('job_skills')
df_ds_explode
```

```python
df_ds_pivot = df_ds_explode.pivot_table(index='job_posted_month', columns='job_skills', aggfunc='size', fill_value=0)
df_ds_pivot.loc['Total'] = df_ds_pivot.sum()
jobs_demand_ds = df_ds_pivot.loc['Total'].sort_values(ascending=False).index
df_ds_pivot = df_ds_pivot[jobs_demand_ds]
df_ds_pivot.drop('Total', inplace=True)
```

```python
df_ds_pivot_top5 = df_ds_pivot.iloc[:,:5].copy()
df_ds_pivot_top5.reset_index(inplace=True)
df_ds_pivot_top5['job_posted_month'] = df_ds_pivot_top5['job_posted_month'].apply(lambda x: pd.to_datetime(x, format="%m").strftime("%B"))
df_ds_pivot_top5.set_index('job_posted_month', inplace=True)
```

```python
df_ds_pivot_top5.plot(
    kind='line',
    linewidth=3,
    linestyle="--",
    marker='o',
    markersize=7,
    colormap='viridis',
    figsize=(10,5)
)
```

```python
df_da = df[df['job_title_short'] == 'Data Analyst']
df_da_exploded = df_da.explode('job_skills')
# df_da_exploded['job_skills'].value_counts()
df_da_pivot = df_da_exploded.pivot_table(index="job_skills", aggfunc="size")
df_da_pivot.sort_values(ascending=False, inplace=True)
df_da_dataframe = df_da_pivot.reset_index(name='skill_count')
df_da_dataframe.shape
```

```python
salary_total = df_da_exploded.groupby('job_skills')['salary_year_avg'].median()
df_da_median_salaries = salary_total.reset_index()
df_da_median_salaries.shape
```

```python
df_da_merged = pd.merge(df_da_dataframe, df_da_median_salaries, how='left', on='job_skills').set_index('job_skills')
df_da_merged
```

```python
df_da_top10 = df_da_merged.head(20).copy().index
df_da_top10_rows = df_da_merged.loc[df_da_top10]
df_da_top10_rows
```

```python
# fig, ax = plt.subplots()
from adjustText import adjust_text

df_da_top10_rows.plot(kind='scatter', x='skill_count', y='salary_year_avg')
texts = []
for i, txt in enumerate(df_da_top10_rows.index):
    texts.append(plt.text(df_da_top10_rows['skill_count'].iloc[i], df_da_top10_rows['salary_year_avg'].iloc[i], txt))
adjust_text(texts, arrowprops=dict(arrowstyle='->', color='red'))
ax = plt.gca()
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f"${int(y/1000)}K"))
plt.show()
```

```python
df_da_india = df[(df["job_country"] == 'India') & (df['job_title_short'] == 'Data Analyst')].copy()
df_da_ind_explod = df_da_india.explode('job_skills')
df_da_ind_explod.dropna(subset='salary_year_avg', inplace=True)
df_grouped = df_da_ind_explod.groupby('job_skills')['salary_year_avg'].agg(['count', 'median'])
df_grouped_salary = df_grouped.sort_values('median', ascending=False).head(10)
df_grouped_volume = df_grouped.sort_values('count', ascending=False).head(10).sort_values('median', ascending=False)
```

```python
df_grouped_salary
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(2,1)
df_grouped_salary.plot(kind='barh', ax=ax[0])
df_grouped_volume.plot(kind='barh', ax=ax[1])
ax[0].invert_yaxis()
ax[1].invert_yaxis()
ax[1].set_xlim(ax[0].get_xlim())
ax[0].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"${int(x/1000)}K"))
ax[1].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"${int(x/1000)}K"))
plt.show()
```

```python
from datasets import load_dataset
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = load_dataset("lukebarousse/data_jobs")['train'].to_pandas()
df_ind = df[df['job_country'] == 'United States'].copy()
df_ind.dropna(subset='salary_rate', inplace=True)
salary_year_avg_median = df_ind['salary_year_avg'].median()
df_ind.fillna({'salary_year_avg': salary_year_avg_median}, inplace=True)
#df_ind['salary_year_avg'].fillna(value=salary_year_avg_median, inplace=True)
jobs = ['Data Scientist', 'Data Engineer', 'Data Analyst']
df_ind_jobs = df_ind[df_ind['job_title_short'].isin(jobs)].copy()
sns.boxplot(data=df_ind_jobs, x='salary_year_avg', y='job_title_short')
plt.xlim(0,600000)
plt.ylabel("")
plt.xlabel("Average salary")
ax = plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x,_: f"${int(x/1000)}K"))
plt.show()
```
